```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
school_grants = pd.read_csv('userssharedsdfschoolimprovement2010grants.csv')
df= pd.DataFrame(school_grants)
print(df)
```

```python
school_grants.columns = ['School Name', 'City', 'State', 'District Name', 'Award_Amount', 'Model Selected', 'Location']
```

```python
school_grants['Award_Amount'] = school_grants.Award_Amount.str.replace('$', '').astype(float)
```

```python
print(school_grants)
```

```python
df_state_average = pd.DataFrame(school_grants.groupby('State').Award_Amount.mean())
df_state_average.sort_values(by=['Award_Amount'], axis=0, ascending=True, inplace=False, kind='quicksort', na_position='last')
```

```python
df_state_average.plot.bar(x=None, y=None)
```

```python
df_ms_state_average = pd.DataFrame(school_grants[school_grants.State =='MS'].head())
print(df_ms_state_average)
```

```python
df_az_state_average = pd.DataFrame(school_grants[school_grants.State =='AZ'].head())
print(df_az_state_average)
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.boxplot(df_ms_state_average['Award_Amount'])
plt.ylabel('Amount')
plt.title('Mississippi')

plt.subplot(1, 2, 2)
plt.boxplot(df_az_state_average['Award_Amount'])
plt.ylabel('Amount')
plt.title('Arizona')
plt.show()
```

```python
df_count = pd.DataFrame(df.groupby('State').City.count())
print(df_count)
```

```python
plt.hist(df_count['City']) 
plt.show()
```

```python
# First plot shows the average amount of grants per state. We can see the award amount varies greatly between states.
# Second group of subplots compare the range of values between Arizona, the highest earning, with Mississippi, the lowest earing
# The final histogram shows how many cities in each state received financial aid. Most states had a low number of cities.
```
