```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd  
%matplotlib inline
```

```python
exponential_var = np.random.exponential(500, 100)
```

```python
mean = np.mean(exponential_var)
sd = np.std(exponential_var)

plt.hist(exponential_var)
plt.axvline(x=mean,color='black')
plt.axvline(x=mean+sd, color='red')
plt.axvline(x=mean-sd, color='red')
plt.show()
```
