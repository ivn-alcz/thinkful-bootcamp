```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
testing = pd.read_csv('states_all.csv')
testing['average_reveneue']=testing['TOTAL_REVENUE']/testing['GRADES_ALL_G']
testing['average_expenditure']=testing['TOTAL_EXPENDITURE']/testing['GRADES_ALL_G']
testing.dtypes
clean_test=testing.dropna()
print(clean_test)
```

```python
clean_test.groupby('STATE')['TOTAL_REVENUE'].aggregate(np.max)
```
